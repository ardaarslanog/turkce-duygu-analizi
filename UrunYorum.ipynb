```python
!pip install transformers

from transformers import pipeline

model_name = "savasy/bert-base-turkish-sentiment-cased"

nlp = pipeline("sentiment-analysis", model=model_name)

yorumlar = [
    "Bu telefon harika, bayıldım!",
    "Ürün çok kalitesiz ve bozuldu.",
    "Kesinlikle tavsiye etmiyorum.",
    "Çok memnun kaldım, teşekkürler.",
    "Hayal kırıklığına uğradım."
]

sonuclar = nlp(yorumlar)

for yorum, sonuc in zip(yorumlar, sonuclar):
    print(f"Yorum: {yorum}")
    print(f"Tahmin: {sonuc['label']} (Score: {sonuc['score']:.2f})\n")
```
